## MATPLOTLIB

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Gráfico de linea
    plt.plot()


```python
#grafico de Linea .plot()
#en este caso visualizaremos una funcion exponencial que aumenta el doble de y por cada x
x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25] 

#dibuja el grafico
plt.plot(x, y,) 
#marca los puntos en el grafico
plt.scatter(x, y,)

plt.xlabel("eje X")
plt.ylabel("eje Y")
plt.title("funcion exponencial")
plt.show() 
```

### Gráfico de barras 
    plt.bar()

```python


categorias = ["A","B","C","D"]
valores = [10,50,60,70]
plt.bar(categorias, valores)
plt.title("grafico de barras")
plt.xlabel("categorias")
plt.ylabel("valores")
plt.show()
```
