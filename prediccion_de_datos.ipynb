## PREDECIR DATOS CON POLINOMIOS


### Predicción de la producción de un país para el año 2000. con datos historicos

```python
import numpy as np
import matplotlib.pyplot as plt

#datos de la produccion de un pais. donde x son los años e y la produccion
x=np.array([1920,1930,1940,1950,1960,1970,1980,1990])
y=np.array([106.46,123.08,132.12,152.27,180.67,205.05,227.23,249.46])

anno = 2000 #año a prededecir
grado = 3 #grado aleatorio de ejemplo, la eleccion del grado se vera mas adelante
coef = np.polyfit(x,y,grado) #polinomio que toma los datos por la potencia del grado que se elija para su prediccion  
p = np.polyval(coef,anno)
print(f'Para el ano {anno} la prediccion es: {p}')

#elegir el grado es muy importante. se debe elegir el grado que menor error en los puntos tenga. el grado es el exponente del polinomio 
# grado 0 = recta horizontal
# grado 1 = recta vertical
# grado 2 = parabola 
# etc etc etc
```

### Visualización de los distintos grados elegidos
    aqui vemos como con cada grado de potencia al polinomio la prediccion cambia por eso debemos verificar cual es el que mas se asemeja a la realidad a travez de los graficos

```python
#funcion para el bucle de graficos con distitnos coeficientes

def fx (x1, coef):
    """
    Evalúa un polinomio representado por sus coeficientes en un conjunto de valores x1.

    Args:
    x1 (array-like): Array de valores donde se evaluará el polinomio.
    coef (list or array-like): Coeficientes del polinomio en orden descendente de grado.

    Returns:
    array-like: Array con los valores resultantes de la evaluación del polinomio en los puntos x1.
    """
    fx = 0
    n = len(coef) - 1
    for p in coef:
        fx = fx + p*x1**n
        n = n - 1
    return fx

#prueba con 10 rangos para ver la diferencia de prediccion al aumentar el coeficiente al polinomio
anno = 2000
for i in range(0,10):
    coef = np.polyfit(x,y,i)
    p = np.polyval(coef, anno)
    
    print(f"para grado {i} la predicción es {p}")
    x1 = np.linspace(1920, anno + 1, 1000) #espacia los datos de manera uniforme 
    y1 = fx(x1, coef) # funcion que toma cada valor del bucle para el polinomio
    plt.figure(figsize=[20,10]) # tamao del grafico
    plt.title("Cantidad de litros vs año. Para grado: " + str(i)) #titulo del grafico
    
    plt.scatter(x,y,s=120,c='blueviolet')  # puntos de interseccion
    plt.plot(x1,y1,"--",linewidth=3,color='orange') # linea punteada
    plt.scatter(anno,p,s=200,c='red') # prediccion del numero del bucle
    plt.yticks(range(100,320,20))
    plt.grid("on") #cuadricula del grafico
    ax=plt.gca() #para realizar modificaciones al grafico 
    ax.set_xlabel("$años$")
    ax.set_ylabel("$Cantidad litros$")
    #plt.savefig("img" + str(i)+".jpg", dpi=600)
    plt.show()
```

### Eleccion del grado que de la eficiencia de la prediccion
     podriamos elegir el mejor grado de prediccion a travez de solo la vista grafica, pero es mejor realizar un analisis de error medio cuadratico que automatice y perfeccione este resultado. Ademas agregando mas variables de potencias al analisis para concluir que la produccion no siempre aumenta al aumentar el grado del polinomio

#### Calculando el error medio cuadratico

```python
anno = 2000
grado = np.arange(0, 30 + 1, 1) #de 0 a 30 de 1 en 1
aproxi = np.array([])
y_pred_vec = np.array([])
for i in grado:
    coef = np.polyfit(x,y,i)
    p = np.polyval(coef, anno)
    aproxi = np.append(aproxi, p)
    
    y_pred_vec = np.array([])
    for j in x: 
        y_pred = np.polyval(coef, j)
        y_pred_vec = np.append(y_pred_vec,y_pred)
    MSE = (sum( (y - y_pred_vec)**2 ))/len(y)
    print(f"Para grado {i} El MES es: {MSE}")
plt.figure(figsize=[20,10])
plt.title("Grado del polinomio vs predicción")
plt.plot(grado,aproxi,"--",linewidth=3,color='red')
plt.grid("on")
plt.show()
```
